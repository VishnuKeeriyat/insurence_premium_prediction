```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv(r"/Users/vishnukeeriyat/Documents/MI projects/Data Set/insurance.csv")
```

```python
df.smoker.value_counts()
```

```python
df.sex.value_counts()
```

```python
df.region.value_counts()
```

```python
df['sex'] = df['sex'].apply(lambda x:1 if x=='male' else 0)
```

```python
df['smoker'] = df['smoker'].apply(lambda x:1 if x=='yes' else 0)
```

```python
df
```

```python
df = df.join(pd.get_dummies(df.region,dtype = int)).drop('region', axis =1)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

```python
X = df.drop('charges', axis =1)
y= df['charges']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = RandomForestRegressor(n_jobs = -1)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
rmse = root_mean_squared_error(y_test, y_pred)
```

```python
rmse
```

```python
mae = mean_absolute_error(y_pred, y_test)
```

```python
mae
```

```python
plt.scatter(y_test, y_pred)
plt.plot(np.linspace(0, max(y_test)), np.linspace(0,max(y_test)),color= 'red')
plt.xlabel('Charges')
plt.ylabel('Prediction')
plt.title('Prediciton VS Truth')
```

```python
feature_importances =sorted(zip(model.feature_names_in_, model.feature_importances_), key =lambda x:x[1], reverse =True)
```

```python
plt.figure(figsize = (10,6))
plt.bar([x[0] for x in feature_importances], [x[1] for x in feature_importances])
plt.title('Feature  Importances')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth' : [None, 2, 5],
    'min_samples_split':[2,4,6,8],
    'min_samples_leaf':[1,2,4,6]
}

model = RandomForestRegressor(n_jobs=-1)

grid_search = GridSearchCV(model, param_grid=param_grid, cv=5)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
model = grid_search.best_estimator_
```

```python
model
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```
